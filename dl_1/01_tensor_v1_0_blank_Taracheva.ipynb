1.4.7 Решить матричное уравнение `A*X*B=-C` - найти матрицу X. Где `A = [[-1, 2, 4], [-3, 1, 2], [-3, 0, 1]]`, `B=[[3, -1], [2, 1]]`, `C=[[7, 21], [11, 8], [8, 4]]`.

````python
A = torch.tensor([[-1., 2., 4.],
                  [-3., 1., 2.],
                  [-3., 0., 1.]])
B = torch.tensor([[3., -1.],
                  [2., 1.]])
C = torch.tensor([[7., 21.],
                  [11., 8.],
                  [8., 4.]])
A_inv = torch.inverse(A)
B_inv = torch.inverse(B)
X = -A_inv @ C @ B_inv
X
````

1.4.8 Проверить, является ли система векторов a1 = (3; −3; 0; 7),
a2 = (2; 2; 4; 7), a3 = (1; 2; 3; 4), a4 = (5; −4; 1; 3) линейно зависимой?

````python
import torch
A = torch.tensor([[3., -3., 0., 7.],
                  [2., 2., 4., 7.],
                  [1., 2., 3., 4.],
                  [5., -4., 1., 3.]])
determinant = torch.det(A)

print(f"определитель: {determinant.item():.2f}")
if determinant == 0:
    print("Система линейно зависима")
else:
    print("Система линейно независима")
````

1.4.9 Сгенирировать тензор из 200 случайных целых чисел, нормально распрделенных cо стандартным отклонением $\sigma = 10$ и матожиданием $\mu = 0$. Построить тензор гистограммы с 20 бинами.

````python
t = torch.normal(mean=0.0, std=10.0, size=(200,)).round().to(torch.int)

min_val = t.min().item()
max_val = t.max().item()

hist, bin_edges = torch.histogram(t.float(), bins=20, range=(min_val, max_val))

print("гистограмма:", hist)
print("границы бинов:", bin_edges)
````
